```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
REWARED_PICKLE_PATH = 'pong_sound_rewards.pkl'
```

```python
reward_sums = pickle.load(open(REWARED_PICKLE_PATH, 'rb'))
```

```python
fig, axes = plt.subplots(2, 1, figsize=(7, 7))
axes[0].set_title('with audio')
axes[1].set_title('w\o audio')
_ = axes[0].plot(reward_sums)
_ = axes[0].set_xlabel('episode')
_ = axes[0].set_ylabel('reward')
_ = axes[1].set_xlabel('episode')
_ = axes[1].set_ylabel('reward')
plt.subplots_adjust(hspace=0.5)
```
